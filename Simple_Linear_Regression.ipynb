```python
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
#Create a simple dataset
hours_studied = np.array([2,3,5,7,9,11,16,19,23])
exam_scores = np.array([65,55,45,21,45,67,89,65,96])
```

```python
hours_studied = hours_studied.reshape(-1,1)

#Create a Linera Regression Model
model = LinearRegression()
model.fit(hours_studied,exam_scores)
```

```python
#Perform the predictions

predicted_score = model.predict([[15]])
print("Predicted exam score:" , predicted_score[0])
```

```python
predicted_score = model.predict(hours_studied)
```

```python
#Import libraries for model evaluation
from sklearn.metrics import mean_absolute_error,mean_squared_error, r2_score
import matplotlib.pyplot as plt
```

```python
mae = mean_absolute_error(exam_scores,predicted_score)
mse = mean_squared_error(exam_scores,predicted_score)
r2 = r2_score(exam_scores, predicted_score)

print("Mean Absolute Error:",mae)
print("Mean Squared Error:",mse)
print("R2 Score:",r2)
```

```python
plt.scatter(hours_studied,exam_scores, label = "Actual Scores")
plt.plot(hours_studied,predicted_score, color = "red", label = "Predicted Scores")
plt.xlabel("Hours Studied")
plt.ylabel("Exam Scores")
plt.legend()
plt.show()
```
