# Calculator

```python
x = float(input("Enter num1: "))
y = float(input("Enter num2: "))
z = input("Enter operator: ")
if z == '+':
  print("Output: ",x+y)
elif z == '-':
  print("Output: ",x-y)
elif z == '*':
  print("Output: ",x*y)
elif z == '/':
  print("Output: ",x/y)
```
