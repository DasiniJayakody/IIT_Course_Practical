# Budget Calculator

```python
# Define the income
income = float(input("Enter your monthly income: "))

# Define the expenses
rent = float(input("Enter rent amount: "))
transport = float(input("Enter Transport amount: "))
entertainment = float(input("Enter entertainment amount: "))
other_expenses = float(input("Enter other expenses : "))
total_expenses = rent + transport + entertainment + other_expenses
#print("Total Expenses:" , total_expense)
savings = income - total_expenses

print("My month income: ", income)
print("My month expense: ", total_expenses)

print("My monthly savings:", savings)
```

```python
# Give a feedback
if savings > 0:
  print("Great ! upi have some savings for this month!")
elif savings == 0:
  print("You broke even this month")
else:
  print("Warning! You are overspending!")
```
